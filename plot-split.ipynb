### Calculate weighted interval score (wis)

```python
q_list = [0.010, 0.025, 0.050, 0.100, 0.150, 0.200, 0.250, 0.300, 0.350, 0.400,
          0.450, 0.500, 0.550, 0.600, 0.650, 0.700, 0.750, 0.800, 0.850, 0.900, 0.950, 0.975, 0.990]
    
# all means, all qs
def cal_quantile_forecasts_all(mean_list, std): 
    all_res = {}
    for q in q_list:
        res = []
        for mean in mean_list:
            val = mean + std * math.sqrt(2) * erfinv(2 * q - 1)
            res = res + [val]
        all_res[q] = np.array(res)
    return all_res
```

```python
cases_quantile_all_1w = cal_quantile_forecasts_all(cases_pred_1w * 120000, np.std(e_cases_1w))
deaths_quantile_all_1w = cal_quantile_forecasts_all(deaths_pred_1w * 500, np.std(e_deaths_1w))
hospitalized_quantile_all_1w = cal_quantile_forecasts_all(hospitalized_pred_1w * 2000, np.std(e_hospitalized_1w))

cases_quantile_all_2w = cal_quantile_forecasts_all(cases_pred_2w * 120000, np.std(e_cases_2w))
deaths_quantile_all_2w = cal_quantile_forecasts_all(deaths_pred_2w * 500, np.std(e_deaths_2w))
hospitalized_quantile_all_2w = cal_quantile_forecasts_all(hospitalized_pred_2w * 2000, np.std(e_hospitalized_2w))

cases_quantile_all_3w = cal_quantile_forecasts_all(cases_pred_3w * 120000, np.std(e_cases_3w))
deaths_quantile_all_3w = cal_quantile_forecasts_all(deaths_pred_3w * 500, np.std(e_deaths_3w))
hospitalized_quantile_all_3w = cal_quantile_forecasts_all(hospitalized_pred_3w * 2000, np.std(e_hospitalized_3w))

cases_quantile_all_4w = cal_quantile_forecasts_all(cases_pred_4w * 120000, np.std(e_cases_4w))
deaths_quantile_all_4w = cal_quantile_forecasts_all(deaths_pred_4w * 500, np.std(e_deaths_4w))
hospitalized_quantile_all_4w = cal_quantile_forecasts_all(hospitalized_pred_4w * 2000, np.std(e_hospitalized_4w))
```
